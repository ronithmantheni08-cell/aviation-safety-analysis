```python
import pandas as pd
import os

# --- The corrected file path and filename ---
file_location = '/Users/ronithmantheni/Documents/python_files'
file_name = 'airport_data.csv'
full_path = os.path.join(file_location, file_name)


# STEP 1: LOAD THE DATA (with the ENCODING FIX)
# ---------------------------------------------------
try:
    # We add encoding='latin1' to handle special characters in the file
    df = pd.read_csv(full_path, encoding='latin1')
    print(f"✅ Step 1: File loaded successfully from: {full_path}")
except FileNotFoundError:
    print(f"❌ FATAL ERROR: The file was not found at {full_path}")
    exit()


# STEP 2: HANDLE MISSING VALUES (Updated to avoid warnings)
# ---------------------------------------------------------
df['Aircraft Out of Service'] = df['Aircraft Out of Service'].fillna(0)
df['Cost Other Infl Adj'] = df['Cost Other Infl Adj'].fillna(0)
df['Cost Repairs Infl Adj'] = df['Cost Repairs Infl Adj'].fillna(0)

for col in ['Precip', 'Sky', 'Species', 'Size', 'Time Of Day', 'Damage', 'Effect']:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')
print("✅ Step 2: Missing values handled.")


# STEP 3: CORRECT DATA TYPES
# --------------------------
# Upgraded function to handle invalid text like 'Feb'
def clean_bird_count(count):
    count = str(count)
    if '-' in count:
        try:
            # Try to split and convert to numbers
            low, high = map(int, count.split('-'))
            return (low + high) / 2
        except ValueError:
            # If it fails (e.g., trying to convert 'Feb'), return a null value
            return None
    # errors='coerce' will turn anything that isn't a valid number into NaN
    return pd.to_numeric(count, errors='coerce')

df['Birds Seen'] = df['Birds Seen'].apply(clean_bird_count)
df['Birds Struck'] = df['Birds Struck'].apply(clean_bird_count)


# The rest of the cleaning functions
def convert_mass(mass_str):
    if isinstance(mass_str, str):
        mass_str = mass_str.replace(',', '').replace(' kg or less', '').replace('above ', '').replace('-', ' ')
        parts = mass_str.split()
        if len(parts) > 1 and parts[1].isdigit():
            return (int(parts[0]) + int(parts[1])) / 2
        elif parts[0].isdigit():
            return int(parts[0])
    return None

df['Aircraft Mass'] = df['Aircraft Mass'].apply(convert_mass)

try:
    df['Incident Date'] = pd.to_datetime(df['Incident Date'], errors='coerce')
except (ValueError, TypeError):
     df['Incident Date'] = pd.to_datetime('1899-12-30') + pd.to_timedelta(df['Incident Date'].astype(float), 'D')

print("✅ Step 3: Data types corrected.")


# STEP 4: CLEAN CATEGORICAL DATA
# ------------------------------
df['Precip'] = df['Precip'].replace('"Fog, Rain"', 'Fog and Rain')
df['Species'] = df['Species'].str.replace(' - small| - medium| - large', '', regex=True)
print("✅ Step 4: Categorical data cleaned.")


# STEP 5: REMOVE DUPLICATES
# -------------------------
duplicates_before = df.duplicated().sum()
df.drop_duplicates(inplace=True)
print(f"✅ Step 5: Removed {duplicates_before} duplicate rows.")


# STEP 6: SAVE THE CLEANED DATA
# -----------------------------
output_filename = os.path.join(file_location, 'cleaned_airport_data.csv')
df.to_csv(output_filename, index=False)
print(f"\n🚀 All done! Your cleaned data has been saved to: {output_filename}")

# Display the info of the final cleaned DataFrame
print("\nFinal Cleaned Data Info:")
df.info()
```
